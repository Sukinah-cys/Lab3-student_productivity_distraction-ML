# **Student Productivity & Distraction Dataset**

**File Name**: student_productivity_distraction_dataset_20000.csv

This dataset contains 20,000 student records with comprehensive information about academic habits, lifestyle factors, digital distractions, and productivity levels. It is designed to support data analysis and machine learning tasks related to student performance and behavioral patterns.

Each row represents an individual student, while each column corresponds to a specific feature that may influence productivity.

### **Key Features**
* study_hours_per_day — Average daily study time
* sleep_hours — Daily sleep duration
* phone_usage_hours — Daily mobile phone usage
* social_media_hours — Time spent on social media platforms
* gaming_hours — Daily time spent on gaming
* stress_level — Reported stress level (scale 1–10)
* focus_score — Measured concentration level (0–100)
* attendance_percentage — Class attendance rate
* final_grade — Academic performance grade
* productivity_score — Overall productivity score (0–100), used as the target variable

### **Dataset Characteristics**
* **Total Records**: 20,000
* **Total Features**: 18
* **Missing Values**: None
* **License**: CC0 Public Domain

### **Recommended Use Cases**
This dataset is suitable for a wide range of analytical and machine learning applications, including:

* Regression modeling for productivity prediction
* Academic performance analysis
* Behavioral and lifestyle impact studies
* Educational data mining
* Data science and machine learning projects
